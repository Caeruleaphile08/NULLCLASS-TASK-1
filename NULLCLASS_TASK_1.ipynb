```python
import pandas as pd
import os

english = data['1'].tolist()
float_en = next(i for i, item in enumerate(english) if isinstance(item, float))
english[float_en] = str(english[float_en])
english = " ".join(english)

french = data['2'].tolist()
float_fr = next(i for i, item in enumerate(french) if isinstance(item, float))
french[float_fr] = str(french[float_fr])
french = " ".join(french)
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSeq2SeqLM

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-fr")

# Initialize an empty list to store translated sentences
translated_sentences = []

# Translate each sentence in the dataset
for sentence in french:
    # Tokenize and translate the sentence
    tokenized_sentence = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
    translated_output = model.generate(**tokenized_sentence, num_beams=4, max_length=50, early_stopping=True)
    translated_sentence = tokenizer.decode(translated_output[0], skip_special_tokens=True)
    translated_sentences.append(translated_sentence)

# Add the translations to your dataset
data['en_translated'] = translated_sentences

# Save the updated dataset
data.to_csv("/content/gdrive/MyDrive/translated_dataset.csv", index=False)
print("Saved")
```
